## Lines in the plane

Maximum number of space $L_n$ given $k$ lines 

When n = 1: space = 2

When n = 2: space = 4

When n = 3: space = 7 

...
$$
L_n = L_{n - 1} + n 
$$

Each new line adds $n$ more new space

$$
\begin{aligned}
L_n &= L_{n-1} + n \\
&= L_{n-2} + n - 1 + n \\
&= L_{n-3} + n - 2 + n - 1 + n \\
&= \dots \\ 
&= L_0 + 0 + 1 + 2 + 3 + \dots + n - 1 + n\\ 
&= L_0 + \frac{(n + 1)n}{2}
\end{aligned}
$$

### What if using bent lines


Each bent line is equal to two normal lines, expect it lacks $2n$ spaces that the extended lines provide

Therefore, 
$$
L_n = \frac{(2n + 1)(2n)}{2} - 2n
$$
